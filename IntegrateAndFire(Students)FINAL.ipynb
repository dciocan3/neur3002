```python
# Jupyter: enable plot rendering inside the notebook (this allows us to actually show the output of the code visually)
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# --- Neuron Parameters (realistic, in mV and pA) ---
#NEURON 1 
C1 = 200.0 #capacitance     
V_th1 = -55 #threshold voltage to produce AP
V_reset1 = -65 #voltage the cell resets to before throwing another AP
V_rest1 = -70 #resting membrane potential
I_ext1 = 300.0  #current flowing into the neuron  
dt = 0.1   #change in time in ms

#NEURON 2
C2 = 200.0 #capacitance     
V_th2 = -55 #threshold voltage to produce AP
V_reset2 = -90 #voltage the cell resets to before throwing another AP
V_rest2 = -50 #resting membrane potential
I_ext2 = 150  #current flowing into the neuron

T = 100          # total simulation time in ms
dt = 0.1         # time step in ms
time = np.arange(0, T, dt)  # time vector

# --- Initialize voltage and spike times, NEURON 1 ---
V_1 = np.full_like(time, V_rest1)  # set all values to resting voltage
spike_times1 = []

# --- Initialize voltage and spike times, NEURON 1 ---
V_2 = np.full_like(time, V_rest2)  # set all values to resting voltage
spike_times2 = []

# --- Simulation Loop ---
for i in range(1, len(time)):
    dV1 = (I_ext1 / C1) * dt       # V = (I / C) * dt, in mV
    V_1[i] = V_1[i-1] + dV1          

    if V_1[i] >= V_th1:            
        V_1[i-1] = 40.0           # mark spike as +40 mV (for visualization)
        V_1[i] = V_reset1         
        spike_times1.append(time[i])

for i in range(1, len(time)):
    dV2 = (I_ext2 / C2) * dt       # V = (I / C) * dt, in mV
    V_2[i] = V_2[i-1] + dV2          

    if V_2[i] >= V_th2:            
        V_2[i-1] = 40.0           # mark spike as +40 mV (for visualization)
        V_2[i] = V_reset1         
        spike_times2.append(time[i])

# --- Plotting ---
plt.figure(figsize=(10, 4))
plt.plot(time, V_1, label='Neuron 1 spiking output')
plt.axhline(V_th1, color='r', linestyle='--', label='threshold')
#plt.plot(time, V_2, label='Neuron 2 spiking output')
plt.xlabel('time (ms)') #what should the x be?
plt.ylabel('Vm (mV)') #what should the y be?
plt.title('Spiking Output') #what is a good title?
plt.legend() #can you think of a legend?
plt.grid(True)
plt.show()
```
